```python
import nltk
emma = nltk.corpus.gutenberg.words('austen-emma.txt')
```

```python
raw_text = emma[11:10000]
```

```python
raw_text = ' '.join(raw_text)
```

```python
length = 10

sequences = []
for i in range(length , len(raw_text)):
    sequences.append(raw_text[i-length:i])
```

```python
sequences
```

```python
chars = sorted(set(raw_text))
```

```python
mapping = {}

for (i , c) in enumerate(chars):
    mapping[c] = i
```

```python
encoded_seqs = []

for seqs in sequences:
    encoded_seqs.append([mapping[c] for c in seqs])
```

```python
import numpy as np
encoded_seqs = np.array(encoded_seqs)
X , y = encoded_seqs[:,:-1], encoded_seqs[:,-1]
```

```python
vocab_size = len(chars)
```

```python
from keras.utils import to_categorical
```

```python
X = to_categorical(X)
```

```python
y = to_categorical(y)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(75, input_shape = (X.shape[1], X.shape[2])))
model.add(Dense(vocab_size, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X , y, epochs=150, verbose = 2)
```

```python
model.save('Character Based NLM.h5')
```

```python
def predict(model, in_text, mapping):

    for _ in range(100):
        encoded = [mapping[c] for c in in_text]
        encoded = encoded[-9:]
        encoded = to_categorical(encoded, num_classes=len(mapping))
        encoded = encoded.reshape(1, encoded.shape[0], encoded.shape[1])
        yhat = model.predict_classes(encoded, verbose=0)
        
        for char, index in mapping.items():
            if index == yhat:
                out_char = char
                break
                
        out_text = in_text + out_char
        in_text = out_text
        
    return out_text
    
```

```python
text = "Between t"
pre = predict(model, text, mapping)
```
